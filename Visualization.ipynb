```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('AAPL.csv')
df
```

*   Date: Ngày giao dịch.
*   Open: Giá mở cửa.
*   High: Giá cao nhất trong ngày.
*   Low: Giá thấp nhất trong ngày.
*   Close: Giá đóng cửa.
*   Volume: Khối lượng giao dịch.
*   Adj Close: Giá đóng cửa điều chỉnh.

```python
import plotly.express as px
from plotly.graph_objs import *
```

```python
open = px.line(df,x='Date',y='Open')
open.update_layout(paper_bgcolor="rgba(0,0,0,0)", plot_bgcolor="rgba(0,0,0,0)",width=1400, height=800)
open.show()
```

```python
High = px.line(df,x='Date',y='High')
High.update_layout(paper_bgcolor="rgba(0,0,0,0)", plot_bgcolor="rgba(0,0,0,0)",width=1400, height=800)
High
```

```python
low = px.line(df,x='Date',y='Low')
low.update_layout(paper_bgcolor="rgba(0,0,0,0)", plot_bgcolor="rgba(0,0,0,0)",width=1600, height=800)
low
```

```python
close = px.line(df,x='Date',y='Close')
close.update_layout(paper_bgcolor="rgba(0,0,0,0)", plot_bgcolor="rgba(0,0,0,0)",width=1400, height=800)
close
```

```python
vol = px.line(df,x='Date',y='Volume')
vol.update_layout(paper_bgcolor="rgba(0,0,0,0)", plot_bgcolor="rgba(0,0,0,0)",width=1400, height=800)
vol
```

```python
adj_close=px.line(df,x='Date',y='Adj Close')
adj_close.update_layout(paper_bgcolor="rgba(0,0,0,0)", plot_bgcolor="rgba(0,0,0,0)",width=1400, height=800)
adj_close
```
